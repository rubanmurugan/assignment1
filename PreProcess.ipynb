```python
import pandas as pd
```

```python
df = pd.read_csv("traffic_stops - traffic_stops_with_vehicle_number.csv")
```

```python
df["search_type"] = df["search_type"].fillna("Unknown")
df
```

```python
df["is_arrested"] = df["is_arrested"].astype(int)
df["search_conducted"] = df["search_conducted"].astype(int)
df["drugs_related_stop"] = df["drugs_related_stop"].astype(int)
```

```python
df["stop_date"] = pd.to_datetime(df["stop_date"], errors='coerce')
df["stop_date"] = df["stop_date"].dt.strftime('%Y-%m-%d')
```

```python
df.to_csv('cleaned_traffic_stops', index=False)
```
